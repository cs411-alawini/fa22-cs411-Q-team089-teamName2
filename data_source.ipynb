```python
import pandas as pd
import numpy as np
import random
```

```python
name = []
for i in range(250):
    name.append("Work")
for i in range(250):
    name.append("Homework")
for i in range(250):
    name.append("Groceries")
for i in range(250):
    name.append("Gym")
    
checklistTable = pd.DataFrame({
    "checkListId": np.arange(1, 1001),
    "names": name,
    "userId": np.arange(1, 1001)
})
checklistTable.head()
```

```python
checklistTable.to_csv("checklistTable.csv")
```

```python
taskContent = []
for i in range(250):
    taskContent.append("Attend 9:30 am meeting")
for i in range(250):
    taskContent.append("Finish Math 100 homework")
for i in range(250):
    taskContent.append("Get eggs")
for i in range(250):
    taskContent.append("Upper body")

dateCompleted = []
pastMonth = np.arange(0, 30)
baseDate = np.datetime64('2022-10-20')
for i in range(1000):
    randomDate = baseDate - np.random.choice(pastMonth)
    dateCompleted.append(randomDate)
    
taskTable = pd.DataFrame({
    "taskId": np.arange(1, 1001),
    "checkListId": np.arange(1, 1001),
    "dateCompleted": dateCompleted,
    "taskContent": taskContent
})

taskTable.head()
```

```python
taskTable.to_csv("taskTable.csv")
```

```python
names = [
    "Liam",
    "Noah",
    "Oliver",
    "Elijah",
    "James",
    "William",
    "Benjamin",
    "Lucas",
    "Henry",
    "Theodore",
    "Olivia",
    "Emma",
    "Charlotte",
    "Amelia",
    "Ava",
    "Sophia",
    "Isabella",
    "Mia",
    "Evelyn",
    "Harper"
]
name = []
for i in range(1000):
    name.append(np.random.choice(names))
    
userTable = pd.DataFrame({
    "userId": np.arange(1, 1001),
    "passwords": ["1" for i in range(1000)],
    "names": name,
    "completionRate": [random.uniform(0,1) for i in range(1000)]
})
userTable.head()
```

```python
userTable.to_csv("userTable.csv")
```
